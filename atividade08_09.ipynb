Para a atividade em questão irei realizar uma análise das 5 séries mais assistidas da Netflix em inglês.

Ref: (https://www.mundoconectado.com.br/cultura/netflix-divulga-filmes-e-series-com-mais-visualizacoes-veja-lista-completa/
)

```python
import pandas as pd

dados_series = {
    'Wandinha': 252100000,
    'Stranger Things': 140700000,
    'Dahmer: Um Canibal Americano': 115600000,
    'Bridgerton': 113300000,
    'O Gambito da Rainha': 112800000
}

serie_visualizacoes = pd.Series(dados_series)
```

**Gráficos de Barras**

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')

serie_visualizacoes.plot(kind='bar', color=['blue', 'green'])
plt.title('SÉRIES MAIS VISTAS EM INGLÊS')
plt.xlabel('TÍTULO')
plt.ylabel('VISUALIZAÇÕES')

plt.show()
```

```python

plt.style.use('seaborn')

serie_visualizacoes.plot(kind='barh', color='skyblue')
plt.title('Séries Mais Vistas em Inglês')
plt.xlabel('Visualizações')
plt.ylabel('Título')

plt.show()
```

**Gráfico de Pizza**

Neste gráfico, você pode observar como as visualizações estão distribuídas entre diferentes séries da Netflix. As conclusões possíveis são:

Identificar quais séries têm uma fatia maior do total de visualizações.
Isso ajuda a destacar as séries mais populares.

Compreender a participação relativa de cada série nas visualizações totais.

Observar se algumas séries dominam as visualizações ou se a distribuição é mais equilibrada.

```python
labels = serie_visualizacoes.index
sizes = serie_visualizacoes.values

plt.style.use('seaborn')

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Visualizações por Série')
plt.axis('equal')


plt.show()
```
