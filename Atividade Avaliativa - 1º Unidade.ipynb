
Avaliando os dicionários de dados, percebemos que muitas variáveis apresentam apenas o SIM ou NÃO. Para essas, faremos apenas um mini dicionário e aplicaremos a funçãoo **REPLACE**, que vai substituir os valores encontrados na tabela por aqueles vistos no dicionário.

Link para os dicionarios:

https://portalsinan.saude.gov.br/acidente-por-animais-peconhentos

```python
dados['MCLI_LOCAL'] = dados['MCLI_LOCAL'].replace({1: 'Sim', 2:'Não', 9:'Ignorado'})
```

```python
#usando um laço de repetição para atribuir os valores Sim e Não
colunas_sim_nao = ['MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA', 'CLI_EQUIMO', 'CLI_NECROS', 'CLI_LOCAL_',
                   'MCLI_SIST', 'CLI_NEURO', 'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL',
                   'CLI_OUTR_2', 'CON_SOROTE', 'COM_LOC', 'COM_SECUND', 'COM_NECROS', 'COM_COMPOR',
                   'COM_DEFICT', 'COM_APUTAC', 'COM_SISTEM', 'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC',
                   'COM_CHOQUE', 'DOENCA_TRA']

for coluna in colunas_sim_nao:
  dados[coluna] = dados[coluna].replace({1: 'Sim', 2:'Não', 9:'Ignorado'})


dados
```

Alterando as colunas do tipo DATA

```python

colunas_data = ['DT_NOTIFIC', 'DT_SIN_PRI',
                'DT_INVEST', 'DT_OBITO', 'DT_ENCERRA',
                'DT_DIGITA', 'ANT_DT_ACI']

for coluna in colunas_data:
  dados[coluna] = pd.to_datetime(dados[coluna], format='%Y%m%d', errors='coerce')


dados
```

Nessa etapa tivemos que reescrever todas as codificações do dicionário de dados para criar vários dicionários a partir do dicionário do SINAN.

```python

dados['ANT_TEMPO_'] = dados['ANT_TEMPO_'].replace({1:'0 - 1h',
                                                   2:'1 - 3h',
                                                   3:'3 - 6h',
                                                   4:'6 - 12h',
                                                   5:'12 e 24h',
                                                   6:'24 e +h',
                                                   9:'Ignorado'})

dados['ANT_LOCA_1'] = dados['ANT_LOCA_1'].replace({1 : 'Cabeça',
                                                   2 : 'Braço',
                                                   3 : 'Ante-Braço',
                                                   4 : 'Mão',
                                                   5 : 'Dedo da Mão',
                                                   6 : 'tronco',
                                                   7 : 'Coxa',
                                                   8 : 'Perna',
                                                   9 : 'Pé',
                                                   10 : 'Dedo do Pé',
                                                   99 : 'Ignorado' })

dados['CLI_TEMPO_'] = dados['CLI_TEMPO_'].replace({1 : 'Normal',
                                                   2 : 'Alterado',
                                                   9 : 'Não realizado'})

dados['TP_ACIDENT'] = dados['TP_ACIDENT'].replace({1 : 'Serpente',
                                                   2 : 'Aranha',
                                                   3 : 'Escorpião',
                                                   4 : 'Lagarta',
                                                   5 : 'Abelha',
                                                   6 : 'Outros',
                                                   9 : 'Ignorado' })

dados['ANI_SERPEN'] = dados['ANI_SERPEN'].replace({1 : 'Botrópico',
                                                   2 : 'Crotálico',
                                                   3 : 'Elapídico',
                                                   4 : 'Laquético',
                                                   5 : 'Serpente não peçonhenta',
                                                   9 : 'Ignorado' })

dados['ANI_ARANHA'] = dados['ANI_ARANHA'].replace({1 : 'Foneutrismo',
                                                   2 : 'Loxoscelismo',
                                                   3 : 'Latroectismo',
                                                   4 : 'Outra aranha',
                                                   9 : 'Ignorado' })

dados['ANI_LAGART'] = dados['ANI_LAGART'].replace({1 : 'Lonomia',
                                                   2 : 'Outra lagarta',
                                                   9 : 'Ignorado'})

dados['TRA_CLASSI'] = dados['TRA_CLASSI'].replace({1 : 'Leve',
                                                   2 : 'Moderado',
                                                   3 : 'Grave',
                                                   9 : 'Ignorado'})

dados['EVOLUCAO'] = dados['EVOLUCAO'].replace({1 : 'Cura',
                                               2 : 'óbito por acidente por animais peçonhentos',
                                               3 : 'Óbito por outras causas',
                                               9 : 'Ignorado'})



dados['CS_GESTANT'] = dados['CS_GESTANT'].replace({1 : '1º Trimestre',
                                                   2 : '2º Trimestre',
                                                   3 : '3º Trimestre',
                                                   4 :'Idade gestacional ignorada',
                                                   5 : 'Não',
                                                   6 : 'Não se aplica',
                                                   9 : 'Ignorado' })

dados['CS_RACA'] = dados['CS_RACA'].replace({1 : 'branca',
                                             2 : 'preta',
                                             3 : 'amarela',
                                             4 : 'parda',
                                             5 : 'indígena',
                                             9 : 'Ignorado'})
```

```python
idades = []

for i in dados["NU_IDADE_N"]:
  if i > 4000:
    idades.append(i - 4000)
  elif 4000 > i :
    idades.append(0)
  else:
    idades.append(0)

dados["NU_IDADE_N_"] = idades

dados
```

**Visualização de dados**

Agora, vamos gerar uma tabela que mostre o número de acidentes por data de ocorrência utilizando a função "value_counts"

```python

acidentes_por_dia = dados_final.value_counts('Data do acidente')

acidentes_por_dia
```

```python
import datetime


filtro_data = dados_final['Data do acidente']>='2022-01-01'

acidentes_por_dia_2022 = dados_final[filtro_data].value_counts('Data do acidente')

acidentes_por_dia_2022
```

Nesse gráfico podemos ver o número de acidentes por data de ocorrência em 2022

```python


acidentes_por_dia_2022.plot(figsize=(10,8), color='r', label='Número de acidentes por dia')

plt.xlabel('Data do acidente')
plt.ylabel('Acidentes por dia')
plt.title('Número de acidentes por animais peçonhentos, por data de ocorrência, Brasil, 2022')
plt.legend()

plt.show()
```

Neste gráfico, podemos ver quais são os animais com maior ocorrência de acidentes

```python

sns.catplot(data=dados_final, x='Tipo de acidente', kind='count', aspect=2)
plt.title('Número de acidentes por animais peçonhentos no Brasil em 2022', fontsize=20)
plt.xlabel('Tipo de acidente')
plt.ylabel('Acidentes', fontsize=12)

plt.show()
```

Agora Vamos realizar os ataques de animais por UF's
