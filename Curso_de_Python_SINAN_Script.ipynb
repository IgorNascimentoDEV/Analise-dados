### Recodificação dos dados
Um dos grandes problemas dos arquivos DBF é a codificação das variáveis. Simplesmente abrir o arquivo não traz informações claras, pois acabamos tendo que recorrer ao dicionário de Dados daquele arquivo. Com o Python, conseguimos recodificar as colunas, escrevendo o dicionário de dados para cada uma. É um rabalho que leva algum tempo, porém, precisamos fazê-lo somente uma vez e depois já estará pronto.

Avaliando p dicionário de dados, percebemos que muitas variáveis apresentam apenas o **SIM** ou **NÃo**. Para essas, faremos apenas um mini dicionário e aplicaremos a funçãoo **REPLACE**, que vai substituir ps valores encontrados na tabela por aqueles vistos no dicionário.

Utilizaremos uma coluna como exemplo, e depois aplicamos para as demais.

```python
dados['MCLI_LOCAL'] = dados['MCLI_LOCAL'].replace({1: 'Sim', 2:'Não', 9:'Ignorado'})
```

Agora, criaremos uma lista com as colunas que possuem codificação de **SIM/NÃO**, e aplicaremos uma das funções mais úteis do Python, oloop de repetição chamado **For**

Essa função nos permite aplicar diversas vezes a mesma peração, alterando apenas um parâmetro

```python
colunas_sim_nao = ['MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA', 'CLI_EQUIMO', 'CLI_NECROS', 'CLI_LOCAL_',
                   'MCLI_SIST', 'CLI_NEURO', 'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL',
                   'CLI_OUTR_2', 'CON_SOROTE', 'COM_LOC', 'COM_SECUND', 'COM_NECROS', 'COM_COMPOR',
                   'COM_DEFICT', 'COM_APUTAC', 'COM_SISTEM', 'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC',
                   'COM_CHOQUE', 'DOENCA_TRA']

for coluna in colunas_sim_nao:
  dados[coluna] = dados[coluna].replace({1: 'Sim', 2:'Não', 9:'Ignorado'})


dados
```

Algumas são do tipo DATA, então vamos alterar para este formato

```python
colunas_data = ['DT_NOTIFIC', 'DT_SIN_PRI',
                'DT_INVEST', 'DT_OBITO', 'DT_ENCERRA',
                'DT_DIGITA', 'ANT_DT_ACI']

for coluna in colunas_data:
  dados[coluna] = pd.to_datetime(dados[coluna], format='%Y%m%d', errors='coerce')


dados
```

```python
dados.info()
```

Agora vem a parte mais  trabalhoosa, onde temos que reescrever todas as codificações do dicionário de dados.
Em outras palaavras, criaremos vários dicionários a partir do dicionário do SINAN

```python


dados['ANT_TEMPO_'] = dados['ANT_TEMPO_'].replace({1:'0 - 1h',
                                                   2:'1 - 3h',
                                                   3:'3 - 6h',
                                                   4:'6 - 12h',
                                                   5:'12 e 24h',
                                                   6:'24 e +h',
                                                   9:'Ignorado'})

dados['ANT_LOCA_1'] = dados['ANT_LOCA_1'].replace({1 : 'Cabeça',
                                                   2 : 'Braço',
                                                   3 : 'Ante-Braço',
                                                   4 : 'Mão',
                                                   5 : 'Dedo da Mão',
                                                   6 : 'tronco',
                                                   7 : 'Coxa',
                                                   8 : 'Perna',
                                                   9 : 'Pé',
                                                   10 : 'Dedo do Pé',
                                                   99 : 'Ignorado' })

dados['CLI_TEMPO_'] = dados['CLI_TEMPO_'].replace({1 : 'Normal',
                                                   2 : 'Alterado',
                                                   9 : 'Não realizado'})

dados['TP_ACIDENT'] = dados['TP_ACIDENT'].replace({1 : 'Serpente',
                                                   2 : 'Aranha',
                                                   3 : 'Escorpião',
                                                   4 : 'Lagarta',
                                                   5 : 'Abelha',
                                                   6 : 'Outros',
                                                   9 : 'Ignorado' })

dados['ANI_SERPEN'] = dados['ANI_SERPEN'].replace({1 : 'Botrópico',
                                                   2 : 'Crotálico',
                                                   3 : 'Elapídico',
                                                   4 : 'Laquético',
                                                   5 : 'Serpente não peçonhenta',
                                                   9 : 'Ignorado' })

dados['ANI_ARANHA'] = dados['ANI_ARANHA'].replace({1 : 'Foneutrismo',
                                                   2 : 'Loxoscelismo',
                                                   3 : 'Latroectismo',
                                                   4 : 'Outra aranha',
                                                   9 : 'Ignorado' })

dados['ANI_LAGART'] = dados['ANI_LAGART'].replace({1 : 'Lonomia',
                                                   2 : 'Outra lagarta',
                                                   9 : 'Ignorado'})

dados['TRA_CLASSI'] = dados['TRA_CLASSI'].replace({1 : 'Leve',
                                                   2 : 'Moderado',
                                                   3 : 'Grave',
                                                   9 : 'Ignorado'})

dados['EVOLUCAO'] = dados['EVOLUCAO'].replace({1 : 'Cura',
                                               2 : 'óbito por acidente por animais peçonhentos',
                                               3 : 'Óbito por outras causas',
                                               9 : 'Ignorado'})



dados['CS_GESTANT'] = dados['CS_GESTANT'].replace({1 : '1º Trimestre',
                                                   2 : '2º Trimestre',
                                                   3 : '3º Trimestre',
                                                   4 :'Idade gestacional ignorada',
                                                   5 : 'Não',
                                                   6 : 'Não se aplica',
                                                   9 : 'Ignorado' })

dados['CS_RACA'] = dados['CS_RACA'].replace({1 : 'branca',
                                             2 : 'preta',
                                             3 : 'amarela',
                                             4 : 'parda',
                                             5 : 'indígena',
                                             9 : 'Ignorado'})

```

A coluna Idade é um pouco diferente. O SINAN utiliza uma codificação neste formato: A composição da varável obedece o seguinte critério: primeiro dígito:

1.Hora

2.Dia

3.Mês

4.Ano


Exemplos:
  - 3009 - nove meses
  - 4018 - dezoito anos

Assim, como apenas nos interessa a idade em anos, diminuiremos todos os números por 4000, Caso o resutado seja negativo, deixamos como zero.

```python
idades = []

for i in dados["NU_IDADE_N"]:
  if i > 4000:
    idades.append(i - 4000)
  elif 4000 > i :
    idades.append(0)
  else:
    idades.append(0)

dados["NU_IDADE_N_"] = idades

dados

```

**Código do IBGE para nome de município**

Da mesma maneira que utilizamos um dicionário e a função replace para alterar elementos no DataFrame, faremos o mesmo para transformar os códigos de IBGE em nomes de municípios, utilizando a função map.

Primeiramente, precisamos de uma tabela onde conversão já esteja pronta.

```python
 municipios = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/Curso_SINAN_Udemy/main/dados/populacao%20ibge%206%20municipio%20br.csv')

municipios
```

Porém, para convertê-lo em dicionário, precisamos realizar algumas alterações, como transformar o código do IBGE em tipo texxto (string), deixar apenas duas colunas (IBGE, e Municipio), e setar a coluna IBGE6 como index.

```python
municipios = municipios[['IBGE6', 'Municipio']]

municipios = municipios.set_index('IBGE6')

municipios
```

Agora, basta utilizar a função **to_dict()** e teremos o nosso dicionário

```python
dicionario_municipios = municipios.to_dict()['Municipio']

dicionario_municipios
```

Para trannsformar os dados, agora vamos utilizar a função map. Pode surgir a dúvida: "**Por que aqui utilizamos o map e antes utilizamos o replace? Existe diferença?**"


Respondendo: O map é uma função que, caso o valor não esteja no dicionário, ele se transforma em NaN, fazendo com que percamos aquele dado, porém, ela é 60 vezes mais rápidado que o replace, então ganhamos tempo quando o dicionário possui mais de 5000 valores.

Assim, convertemos as duas colunas que onde temos os códiigos do IBGE(ID_MUNICIP, ID_MN_RESI e ANT_MUNIC_).

```python
dados.info()
```

```python
#criando colunas com o dicionario de municipio
dados['ID_MUNICIP_nome'] = dados['ID_MUNICIP'].map(dicionario_municipios)
dados['ID_MN_RESI_nome'] = dados['ID_MUNICIP'].map(dicionario_municipios)
dados['ANT_MUNIC_nome'] = dados['ID_MUNICIP'].map(dicionario_municipios)

dados
```

Da mesma manneira, é possível alterar a codificação da OCUPAÇÃO

```python
ocupacao = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/Curso_SINAN_Udemy/main/dados/CBO2002%20-%20Ocupacao.csv')
ocupacao = ocupacao.set_index('CODIGO')
dicionario_ocupacao = ocupacao.to_dict()['TITULO']


dados['ID_OCUPA_N_nome'] = dados['ID_OCUPA_N'].map(dicionario_ocupacao)
dados
```

Chegou a hora de alterar também a codificação da UF para a sua respectiiva sigla, da mesma maneira como fizemos para os municípios.

```python
tabela_uf = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/Curso_SINAN_Udemy/main/dados/cod_uf.csv')
tabela_uf = tabela_uf.set_index('Código UF')
dicionario_uf = tabela_uf.to_dict()['UF']


dados['SG_UF_nome'] = dados['SG_UF'].map(dicionario_uf)
dados['SG_UF_NOT_nome'] = dados['SG_UF_NOT'].map(dicionario_uf)
dados['ANT_UF_nome'] = dados['ANT_UF'].map(dicionario_uf)
dados
```
