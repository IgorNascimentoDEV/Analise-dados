```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tencent',
            "McDonald's", 'Visa',
            'Facebook', 'Alibaba',
            'Louis Vuitton']
```

```python

bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526,
           191.032, 186.421, 169.966,
           124.273  ]
```

```python
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel("Empresas")
plt.ylabel('Bilhes')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel("Empresas")
plt.ylabel('Bilhes')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15, 8)
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel("Empresas")
plt.ylabel('Bilhes')
plt.show()
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel('Bilhes', fontsize=20)
```

```python
plt.plot(empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel('Bilhes', fontsize=20)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
```

```python
plt.plot(empresas, bilhoes, color = 'green', lw=8)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel('Bilhes', fontsize=20)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
```

```python
plt.plot(empresas, bilhoes, color = 'green', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel('Bilhes', fontsize=20)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
```

```python
plt.plot(empresas, bilhoes, color = 'black', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel('Bilhes', fontsize=20)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.bar(empresas, bilhoes, color = 'blue')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel('Bilhes', fontsize=20)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
plt.show()
```

```python

cores = ['black', 'dimgray', 'dimgrey', 'gray', 'grey',
         'darkgray', 'darkgrey', 'silver', 'lightgray', 'lightgrey']
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel('Bilhes', fontsize=20)
plt.grid(True)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-white')
plt.bar(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel('Bilhes', fontsize=20)
plt.grid(True)
plt.show()
```

```python
plt.style.use('seaborn-dark')
plt.barh(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.ylabel("Empresas", fontsize=20)
plt.xlabel('Bilhes', fontsize=20)
plt.grid(True)
plt.show()
```

```python
plt.style.use('seaborn-dark')
plt.scatter(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.ylabel("Empresas", fontsize=20)
plt.xlabel('Bilhes', fontsize=20)
plt.grid(True)
plt.show()
```
